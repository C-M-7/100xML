```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/train39_dataset.csv')[['Age', 'Pclass', 'Fare', 'Survived']]
df.head()
```

```python
df.isnull().mean()
```

```python
x = df.drop(['Survived'], axis=1)
y = df['Survived']
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 2)
```

## using KNNImputer for missing values in x_train,x_test

```python
knn = KNNImputer(n_neighbors = 3, weights = 'distance')

x_train_trf = knn.fit_transform(x_train)
x_test_trf = knn.transform(x_test)
```

```python
pd.DataFrame(x_train_trf, columns = x_train.columns)
```

## applying model on KNN-imputed data

```python
lr = LogisticRegression()
lr.fit(x_train_trf, y_train)
y_pred = lr.predict(x_test_trf)
accuracy_score(y_test, y_pred)
```

## using SimpleImputer for comparison

```python
si = SimpleImputer()
x_train_trf2 = si.fit_transform(x_train)
x_test_trf2 = si.transform(x_test)
```

```python
lr = LogisticRegression()
lr.fit(x_train_trf2, y_train)
y_pred = lr.predict(x_test_trf2)
accuracy_score(y_test, y_pred)
```

## clearly the accuracy of KNN imputed values is more than Simple Imputed values
